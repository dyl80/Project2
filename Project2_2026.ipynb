2 (10pts). Next, write a function to evaluate how well the model performs (for this we can just use classification rate as the metric - # of correct classifications).  Note this is a rate so should return the accuracy of your model:
$$
    CR = \frac{\text{Correct Classification}}{\text{All Instances Tested}}
$$

```python
# Function definition here, returns the classification rate between your models predicted target and the true value #
def ClassificationRate(m_model_params, x_input, t_target):
    correct = 0
    score = np.dot(x_input, m_model_params)
    # 0 or 1
    predictions = ( score > 0).astype(int)

    # Decides if 
    for i in range(len(t_target)):
        if predictions[i] == t_target[i]:
            correct += 1
    CR = correct / len(t_target)
    return CR
```

3 (10pts). It is very important when using logistic regression that we scale our features appropriatly, using the standard scalar, write a function to scale the features before fitting the model:

* **Recall:** We compute the mean and deviation of the training set, but then use that mean and deviation on the test set (DO NOT recompute the mean and deviation for the test set)

$$
    \text{StandardScale} = \frac{\textbf{x} - \bar{\textbf{x}}}{\sigma}
$$

```python
# Write your standard_scaler function here #
def SS(x_input):
    ss_x_input = []
    mean = sum(x_input) / len(x_input)
    variance = sum((x - mean)**2 for x in x_input) / len(x_input)
    stdev = variance**0.5

    # Apply scale
    for item in x_input:
        ss_x_input.append( (item - mean) / stdev )
    ss_x_input = np.array(ss_x_input)
    return ss_x_input
```

4 (30pts). Next write a function (called LogisticRegression) to learn the model parameters using gradient descent (or stochastic gradietn descent if you choose). 
    - The function will take as input, the features (x_input), the target values (t_target), and learning rate. (only using the training set!!!!!)
    - The function will return the model parameters.

* **Hint**:  You might consider creating a "helper" function to compute the gradient descent for you.

**Note**  
* You should assume the model takes the form
$$
    y = \sigma \left( w_0 + \sum_{j=1}^M x_j w_j \right) = \sigma (\textbf{w}^T \textbf{x} + w_0),
$$
where $j$ corresponds to the $j^{\text{th}}$ feature, and recall
$$
    \sigma (z) = \frac{1}{1 + \exp(-z)}.
$$

* **Note:** We derived the gradients for the weight updates as:
$$
    \frac{\partial \ell(\textbf{w})}{\partial w_j} = \sum_{n=1}^N \left( x_j^{(n)} \left( t^{(n)} - \underbrace{\frac{1}{1 + \text{exp}(-z^{(n)})}}_{p(t^{(n)}=1|\textbf{x}^{(n)})} \right) \right),
$$
where recall that $x_j^{(n)}$ is the $j^\text{th}$ feature associated with the $n^\text{th}$ training sample.

```python
# Function to compute logistic regression model parameters
def LogisticRegression(t_target, x_input, l_learning_rate):
    # standard scale the features of the training data
    ss_x_input = SS(x_input)
    
    m = ss_x_input.shape[0] # number of data points
    n = ss_x_input.shape[1] # number of features
    w = np.zeros(n) # vector of 0s
    w0 = 0 # AKA b AKA intercept
    tol = 1e-6
    cost_prev = float('inf')

    max_iterations = 10000
    for i in range(max_iterations):
        z = np.matmul(ss_x_input, w) + w0 # get raw score
        pred = 1 / (1 + np.exp(-z)) # squash raw score to a probability using sigmoid function

        # check correctness using binary cross-entropy. added epsilon to prevent log(0).
        cost_curr = (-1/m) * (np.sum(t_target * np.log(pred + 1e-15) + (1 - t_target) * np.log(1 - pred + 1e-15)))
        if abs( cost_prev - cost_curr ) < tol:
            break
        cost_prev = cost_curr

        # calculate gradients
        error = pred - t_target
        dw = (1 / m) * np.dot(ss_x_input.T, error) 
        dw0 = (1 / m) * np.sum(error)

        # update w and w0
        w -= l_learning_rate * dw
        w0 -= l_learning_rate * dw0
        
    # w = [w, w0]
    return w, w0
```

5 (10pts). Evaluate the performance of the classifier.  Similar to $k$-fold cross validation, to get a good idea of how well your model is performing, do an 80/20 split, and run 5-fold validation on the splits.  

**Example:** Do a random 80/20 split.  Using 80% of the data, train your model and test on the 20% remaining.  Do this at least 5 times (grabbing a new random 80/20 split of the data) and compute the mean and devation for the classification accuracy.

* Note: This is NOT $k$-fold, but it will give you a better idea about how well your model is working, etc.

```python
# Model evaluation (display the mean and deviation of the classification rate) #

num_folds = 5
rng = np.random.default_rng(seed=448)

# Shuffle training indices
indices = np.arange(len(x_train))
rng.shuffle(indices)

# Split indices into 5 roughly equal folds
fold_size = len(indices) // num_folds
indices = indices[:fold_size * num_folds]
fold_indices = [indices[i*fold_size:(i+1)*fold_size] for i in range(num_folds)]
accuracies = []
l_rate = 0.01

for i in range(num_folds):
    val_idx = fold_indices[i]
    train_idx = np.concatenate([fold_indices[j] for j in range(num_folds) if j != i])

    x_fold_train, t_fold_train = x_train[train_idx], t_train[train_idx]
    x_fold_val, t_fold_val = x_train[val_idx], t_train[val_idx]

    # Train model
    w, w0 = LogisticRegression(t_fold_train, x_fold_train, l_rate)

    # Predict on validation fold
    ss_x_val = SS(x_fold_val)
    z_val = np.dot(ss_x_val, w) + w0
    pred_probs = 1 / (1 + np.exp(-z_val))
    predictions = (pred_probs >= 0.5).astype(int)

    # Compute accuracy
    acc = np.mean(predictions == t_fold_val)
    accuracies.append(float(acc))  # <- now executed for every fold

mean_acc = np.mean(accuracies)
std_acc = np.std(accuracies)

print(len(predictions), len(t_fold_val))
print("5-Fold CV Accuracy Scores:", accuracies)
print("Mean Accuracy:", mean_acc)
print("Std Deviation:", std_acc)
```

2 (15pts). Assuming a linear regression model, with a softmax output, write a function called LogRegMultClass(t_target, x_input, l_learning_rate) that returns the model paramters.

* **Note:** Similar to the above, you should use an 80/20 split to divide the data into a training vs. testing set.  You might want to think about how you can use your Gradient descent helper function in Problem A to cmpute the gradients for this problem as well.
* **Hint:** You might consider writing a quick helper function to perform the softmax output, and recall:
$$
            \text{softmax}(z_1, \dots, z_k) = \left[
            \begin{array}{c}
                \frac{\exp{(z_1)}}{\sum_{j=1}^k \exp{(z_j)}}\\
                \vdots \\
                \frac{\exp{(z_k)}}{\sum_{j=1}^k \exp{(z_j)}}\\
            \end{array}
            \right]
$$ 

```python
# Perform multiclass logistic regression #
def LogRegMultClass(t_target, x_input, l_learning_rate):
    # standard scale the features of the training data
    ss_x_input = SS(x_input)
    k = len(np.unique(t_target))
    m = ss_x_input.shape[0] # number of data points
    n = ss_x_input.shape[1] # number of features
    w = np.zeros(n, k) # matrix of 0s
    b = np.zeros(k)
    T = np.zeros((m, k))
    T[np.arange(m), t_target] = 1
    tol = 1e-6
    cost_prev = float('inf')

    max_iterations = 10000
    for i in range(max_iterations):
        z = np.matmul(ss_x_input, w) + b # get raw score
        expZ = np.exp( z - np.max(z, axis=1, keepdims=True)) # Softmax
        pred = expZ / np.sum(expZ, axis=1, keepdims=True)

        # check correctness using binary cross-entropy. added epsilon to prevent log(0).
        cost_curr = (-1/m) * np.sum(T * np.log(pred + 1e-15))
        if abs( cost_prev - cost_curr ) < tol:
            break
        cost_prev = cost_curr

        # calculate gradients
        error = pred - T
        dw = (1 / m) * np.dot(ss_x_input.T, error) 
        dw0 = (1 / m) * np.sum(error, axis=0)

        # update w and w0
        w -= l_learning_rate * dw
        b -= l_learning_rate * dw0
        
    return w, b
```
